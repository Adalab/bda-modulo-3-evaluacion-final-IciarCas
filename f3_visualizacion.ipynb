```python
import pandas as pd
import numpy as numpy

pd.set_option("display.max_columns", None)

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_merged = pd.read_csv("data/merged_customer_flight_loyalty.csv")
```

# 3. Visualización

```python
orden_meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
               'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

vuelos_mes = df_merged.groupby('month')['flights_booked'].sum().reindex(orden_meses)

plt.figure(figsize=(12, 5))
sns.barplot(x=vuelos_mes.index, y=vuelos_mes.values)
plt.title('Distribución de vuelos reservados por mes')
plt.xticks(rotation=45)
plt.show()
```

Los datos muestran que los meses en los que más vuelos se reservan son los de verano (mayo-septiembre), así como diciembre. Probablemente es debido a que se aprovecha el buen tiempo de verano y periodos vacacionales en invierno para viajar.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_merged, x='distance', y='points_accumulated', alpha=0.3)
plt.title('Relación entre distancia y puntos acumulados')
plt.show()
```

Hay una clara relación entre la distancia y los puntos acumulados, que se puede apreciar en la tendencia ascendente del gráfico.

```python
clientes_provincia = df_merged['province'].value_counts()

plt.figure(figsize=(12, 5))
sns.barplot(x=clientes_provincia.index, y=clientes_provincia.values)
plt.title('Distribución de clientes por provincia o estado')
plt.xticks(rotation=45)
plt.show()
```

La mayoría de los clientes viven en tres provincias o estados principales: Ontario, British Columbia y Quebec, lo cual es representativo de las áreas más pobladas de Canadá.

```python
salario_educacion = df_merged.groupby('education')['salary'].mean().sort_values()

plt.figure(figsize=(10, 5))
sns.barplot(x=salario_educacion.index, y=salario_educacion.values)
plt.title('Salario promedio por nivel educativo')
plt.xticks(rotation=45)
plt.show()
```

La distribución de salario por grupo educativo muestra que los niveles educativos con mayor salario son los doctorados y a cierta distancia los másteres. Esto nos deja ver que, a mayor nivel de educación alcanzado, más elevado es el salario de los clientes.

```python
plt.figure(figsize=(6, 6))
df_merged['loyalty_card'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Proporción de clientes según tarjeta de fidelidad')
plt.ylabel('')
plt.show()
```

Se puede apreciar que casi la mitad de los clientes cuentan con una tarjeta Star, y un tercio tienen la Nova. Aurora es la tarjeta que menos clientes tiene y supone una quinta parte de los registros.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df_merged, x='marital_status', hue='gender')
plt.title('Clientes por estado civil y género')
plt.show()
```

La mayoría de los usuarios registrados están casados, en igual proporción para ambos géneros. El segundo registro más alto lo componen las personas solteras, también en similar medida para hommbres y mujeres, y y en tercer lugar están los clientes cuyo estado civil es divorciado, de nuevo en una cantidad parecida para ambos géneros.
