# 4. Evaluación de Diferencias en Reservas de Vuelos por Nivel Educativo

```python
df_merged = pd.read_csv("data/merged_customer_flight_loyalty.csv")
```

```python
df_fase4 = df_merged[['flights_booked', 'education']].copy()
```

```python
descriptive_analysis = df_fase4.groupby('education')['flights_booked'].agg(
    media='mean',
    mediana='median',
    desviacion_std='std',
    minimo='min',
    maximo='max',
    count='count'
).round(2)

print(descriptive_analysis)
```
